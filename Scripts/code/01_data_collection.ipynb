## 1. Directory Setup

```python
RAW_DIR = os.path.join("..", "data", "raw")
os.makedirs(RAW_DIR, exist_ok=True)
print(f"Raw data directory ready: {os.path.abspath(RAW_DIR)}")
```

## 2. Helper Utilities

```python
collection_log = []

def md5(filepath):
    h = hashlib.md5()
    with open(filepath, "rb") as f:
        for chunk in iter(lambda: f.read(8192), b""):
            h.update(chunk)
    return h.hexdigest()

def log_dataset(name, source, filepath, method):
    df = pd.read_csv(filepath)
    entry = {
        "dataset"   : name,
        "source"    : source,
        "method"    : method,
        "filepath"  : filepath,
        "rows"      : df.shape[0],
        "columns"   : df.shape[1],
        "size_kb"   : round(os.path.getsize(filepath) / 1024, 2),
        "md5"       : md5(filepath),
        "timestamp" : datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    }
    collection_log.append(entry)
    print(f"  Logged  : {name}")
    print(f"  Shape   : {df.shape[0]} rows x {df.shape[1]} columns")
    print(f"  Size    : {entry['size_kb']} KB")
    print(f"  MD5     : {entry['md5']}")
    print(f"  Saved to: {filepath}")
    return df
```

## 3. Dataset 1 -- PIMA Indians Diabetes Database

**Source:** UCI ML Repository  
**URL:** https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv  
**Method:** Direct URL download via `urllib.request`  
**License:** Public domain / UCI open access


```python
PIMA_URL  = (
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/"
    "pima-indians-diabetes.data.csv"
)
PIMA_COLS = [
    "Pregnancies", "Glucose", "BloodPressure", "SkinThickness",
    "Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"
]
PIMA_PATH = os.path.join(RAW_DIR, "pima_diabetes.csv")

print("Downloading PIMA Indians Diabetes Database ...")
urllib.request.urlretrieve(PIMA_URL, PIMA_PATH)

# The file has no header -- attach column names on save
df_pima = pd.read_csv(PIMA_PATH, header=None, names=PIMA_COLS)
df_pima.to_csv(PIMA_PATH, index=False)

print("Download complete.")
pima_preview = log_dataset(
    name    = "PIMA Indians Diabetes Database",
    source  = PIMA_URL,
    filepath= PIMA_PATH,
    method  = "urllib.request.urlretrieve"
)
pima_preview.head()
```

## 4. Dataset 2 -- UCI Heart Disease Dataset (Cleveland)

**Source:** UCI ML Repository  
**URL:** https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data  
**Method:** Direct URL download via `urllib.request`  
**License:** UCI open access


```python
HEART_URL  = (
    "https://archive.ics.uci.edu/ml/machine-learning-databases/"
    "heart-disease/processed.cleveland.data"
)
HEART_COLS = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
    "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"
]
HEART_PATH = os.path.join(RAW_DIR, "uci_heart_disease.csv")

print("Downloading UCI Heart Disease Dataset ...")
urllib.request.urlretrieve(HEART_URL, HEART_PATH)

df_heart = pd.read_csv(HEART_PATH, header=None, names=HEART_COLS, na_values="?")
df_heart.to_csv(HEART_PATH, index=False)

print("Download complete.")
heart_preview = log_dataset(
    name    = "UCI Heart Disease Dataset (Cleveland)",
    source  = HEART_URL,
    filepath= HEART_PATH,
    method  = "urllib.request.urlretrieve"
)
heart_preview.head()
```

## 5. Dataset 3 -- Framingham Heart Study Dataset

**Source:** Framingham Heart Study (publicly mirrored on GitHub)  
**URL:** https://raw.githubusercontent.com/datasciencedojo/datasets/master/framingham.csv  
**Method:** Direct URL download via `urllib.request`  
**Relevance:** Contains systolic/diastolic BP, cholesterol, BMI, smoking, and glucose with a binary `TenYearCHD` target, making it suitable for cardiovascular and hypertension risk modelling.  
**License:** Public domain -- original study data released for educational use.


```python
FRAM_URL = (
    "https://raw.githubusercontent.com/GauravPadawe/"
    "Framingham-Heart-Study/master/framingham.csv"
)
FRAM_PATH = os.path.join(RAW_DIR, "framingham.csv")

print("Downloading Framingham Heart Study Dataset ...")
urllib.request.urlretrieve(FRAM_URL, FRAM_PATH)

print("Download complete.")
fram_preview = log_dataset(
    name    = "Framingham Heart Study Dataset",
    source  = FRAM_URL,
    filepath= FRAM_PATH,
    method  = "urllib.request.urlretrieve"
)
fram_preview.head()
```

## 6. Collection Log
